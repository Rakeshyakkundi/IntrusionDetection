```python
data = data[["duration","protocol_type","src_bytes","dst_bytes","is_host_login","is_guest_login","diff_srv_rate","srv_diff_host_rate","flag","labels"]]
```

```python
data['protocol_type'].unique()
```

```python
def ret_protocol(a):
    if str(a) == "tcp":
        return 1
    elif str(a) == "udp":
        return 2
    elif str(a) == "icmp":
        return 3
data['protocol_type'] = data['protocol_type'].apply(ret_protocol)
```

```python
data['protocol_type'].unique()
```

Flag

```python
data['flag'].unique()
```

```python
def ret_flag(a):
    if str(a) == "SF":
        return 1
    elif str(a) == "S0":
        return 2
    elif str(a) == "S1":
        return 3
    elif str(a) == "S2":
        return 4
    elif str(a) == "S3":
        return 5
    elif str(a) == "REJ":
        return 6
    elif str(a) == "RSTO":
        return 7
    elif str(a) == "RSTR":
        return 8
    elif str(a) == "RSTOS0":
        return 9
    elif str(a) == "OTH":
        return 10
    elif str(a) == "SH":
        return 11
data['flag'] = data['flag'].apply(ret_flag)
```

```python
data['flag'].unique()
```

```python
data['labels'].unique()
```

```python
def ret_labels(a):
  if str(a)== 'normal.':
    return 1
  elif str(a) == "buffer_overflow.":
    return 2
  elif str(a) == "loadmodule.":
    return 3
  elif str(a) == "perl.":
    return 4
  elif str(a) == "neptune.":
    return 5
  elif str(a) == "smurf.":
    return 6
  elif str(a) == "guess_passwd.":
    return 7
  elif str(a) == "pod.":
    return 8
  elif str(a) == "teardrop.":
    return 9
  elif str(a) == "portsweep.":
    return 10
  elif str(a) == "ipsweep.":
    return 11
  elif str(a) == "land.":
    return 12
  elif str(a) == "ftp_write.":
    return 13
  elif str(a) == "back.":
    return 14
  elif str(a) == "imap.":
    return 15
  elif str(a) == "satan.":
    return 16
  elif str(a) == "phf.":
    return 17
  elif str(a) == "nmap.":
    return 18
  elif str(a) == "multihop.":
    return 19
  elif str(a) == "warezmaster.":
    return 20
  elif str(a) == "warezclient.":
    return 21
  elif str(a) == "spy.":
    return 22
  elif str(a) == "rootkit.":
    return 23
data['labels'] = data['labels'].apply(ret_labels)
```

```python
data.head()
```

Sklearn

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
X = data.drop(columns=["labels"])
Y = data["labels"]
```

```python
train,val_train,test,val_test = train_test_split(X,Y,test_size=0.5,random_state=355)
x_train,x_test,y_train,y_test = train_test_split(train,test,test_size=0.2,random_state=355)
```

KNN

```python
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
```

SVM

```python
svm = SVC()
svm.fit(x_train,y_train)
```

```python
logic = LogisticRegression()
logic.fit(x_train,y_train)
```

```python
print("Knn:",knn.score(x_test,y_test))
print("svm:",svm.score(x_test,y_test))
print("logistic",logic.score(x_test,y_test))
```
